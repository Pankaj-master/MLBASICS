```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.scatter(setosa["sepal length (cm)"],setosa["sepal width (cm)"],label = "Setosa")
plt.scatter(verginica["sepal length (cm)"],verginica["sepal width (cm)"],color = "green",label = "Verginica")
plt.scatter(versi["sepal length (cm)"],versi["sepal width (cm)"],color = "red",label = "Versicolor")
plt.legend()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y = train_test_split(x,y,test_size = 0.3)
```

```python
print(len(test_x))
print(len(train_x))
print(len(train_y))
print(len(test_y))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors = 10)
```

```python
model.fit(train_x,train_y)
```

```python
model.score(test_x,test_y)
```

```python
model.predict(test_x)==test_y
```

```python
test_y
```

```python
for i in range(1,50):
    model = KNeighborsClassifier(n_neighbors = i)
    model.fit(train_x,train_y)
    print("K is :",i,"    :     ",model.score(test_x,test_y))
```
