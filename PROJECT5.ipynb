```python
'''In order to convert string value into numerical value there are two methods:
1. Label Encoding
2. One Hot Encoding'''
#Program to get male or female using height and weight
```

```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/panku/Downloads/kaggle_bmi.csv")
```

```python
df
#decision tree algo beinng used
```

```python
x = df[["Height","Weight"]]
y = df.Gender
```

```python
x
```

```python
y
```

```python
#One Hot Encoding Method -- pd.get_dummies(columnn_name)
Y = pd.get_dummies(y)
```

```python
Y
```

```python
from sklearn.tree import DecisionTreeClassifier
import warnings
warnings.filterwarnings("ignore")
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x,Y)
```

```python
ans = model.predict([[190,60]])
```

```python
ans
```

```python
ans = model.predict([[130,60]])
```

```python
if ans[0][0] == True:
    print("Female")
else:
    print("Male")
```
